#### Exercice 15, chap 3 (p126)

##### (a) 

```r
#zn, indus, chas, nox, rm, age, dis, rad, tax, ptratio, black,lstat, medv

#We fit for each predictor a simple linear regression
boston_zn = lm(crim~zn, Boston)
summary(boston_zn) #p-value < 0.05, so we can reject null hypothesis

boston_indus = lm(crim~indus, Boston)
summary(boston_indus) #p-value < 0.05, so we can reject null hypothesis

boston_chas = lm(crim~chas, Boston)
summary(boston_chas) #p-value > 0.05, so we can't reject null hypothesis

boston_nox = lm(crim~nox, Boston)
summary(boston_nox) #p-value < 0.05, so we can reject null hypothesis

boston_rm = lm(crim~rm, Boston)
summary(boston_rm) #p-value < 0.05, so we can reject null hypothesis

boston_age = lm(crim~age, Boston)
summary(boston_age) #p-value < 0.05, so we can reject null hypothesis

boston_dis = lm(crim~dis, Boston)
summary(boston_dis) #p-value < 0.05, so we can reject null hypothesis

boston_rad = lm(crim~rad, Boston)
summary(boston_rad) #p-value < 0.05, so we can reject null hypothesis

boston_tax = lm(crim~tax, Boston)
summary(boston_tax) #p-value < 0.05, so we can reject null hypothesis

boston_ptratio = lm(crim~ptratio, Boston)
summary(boston_ptratio) #p-value < 0.05, so we can reject null hypothesis

boston_black = lm(crim~black, Boston)
summary(boston_black) #p-value < 0.05, so we can reject null hypothesis

boston_lstat = lm(crim~lstat, Boston)
summary(boston_lstat) #p-value < 0.05, so we can reject null hypothesis

boston_medv = lm(crim~medv, Boston)
summary(boston_medv) #p-value < 0.05, so we can reject null hypothesis
```

##### (b) 

```r
crim_reg = lm(crim~., Boston)
summary(crim_reg)
```

We can see that indus, chas, rm, age, tax, ptratio and lstat have a p_value above 0.05 and are not significant so we must keep h0 hypothesis
